## catplot


```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
#catplot


data=sns.load_dataset("tips") # Load the tips dataset from seaborn's built-in datasets
```

```python
data.head()
```

```python
sns.catplot(x='day',y='total_bill',data=data)
```

```python
# Create a categorical plot (Box Plot) with 'day' on the x-axis and 'total_bill' on the y-axis
sns.catplot(
    x='day',                # Categorical variable (Day of the week: Thursday, Friday, Saturday, Sunday)
    y='total_bill',         # Continuous variable (Total bill amount for the meal)
    data=data,              # Data to plot (the tips dataset)
    kind='box'              # Type of plot: 'box' creates a box plot
)
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='bar')
```

```python
# Create a categorical plot (Box Plot) with 'day' on the x-axis and 'total_bill' on the y-axis
sns.catplot(
    x='day',                # Categorical variable (Day of the week: Thursday, Friday, Saturday, Sunday)
    y='total_bill',         # Continuous variable (Total bill amount for the meal)
    data=data,              # Data to plot (the tips dataset)
    kind='violin'              # Type of plot: 'box' creates a box plot
)
```

```python
# Create a categorical plot (Box Plot) with 'day' on the x-axis and 'total_bill' on the y-axis
sns.catplot(
    x='day',                # Categorical variable (Day of the week: Thursday, Friday, Saturday, Sunday)
    y='total_bill',         # Continuous variable (Total bill amount for the meal)
    data=data,              # Data to plot (the tips dataset)
    kind='boxen'              # Type of plot: 'box' creates a box plot
)
```
